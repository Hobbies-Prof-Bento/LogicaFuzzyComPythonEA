# Lógica Fuzzy com Python: O guia para Iniciantes: Exemplo do aspirador de pó

### Antecedentes (entradas)

Superfície: qual nível de dificuldade da superfície, em uma escala de 0 a 5?
    
- fácil, moderada, difícil
    
Sujeira: qual a quantidade de sujeira, em uma escala de 0 a 5?

- leve, moderada, pesada
    
### Consequente (saída)

Sucção: qual o nível de sucção que o aspirador de pó deve ser configurado, entre 0% e 10%?

- baixa, média, alta
    
regras
    
   - se a superfície for fácil e a sujeira for leve então a sucção será baixa
   - se a superfície for moderada e a sujeira for leve então a sucção será média
   - se a superfície for dificil ou a sujeira for pesada então a sucção será alta
   - se a superfície for moderada então a sucção será média

## instalação e importação das bibliotecas

- Documentação: https://pythonhosted.org/scikit-fuzzy/overview.html

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib
```

## Antecedentes e consequentes

```python
superficie = ctrl.Antecedent(np.arange(0,6,1),'superficie')
sujeira = ctrl.Antecedent(np.arange(0,6,1),'sujeira')
```

```python
superficie
```

```python
superficie.universe
```

```python
sujeira
```

```python
sujeira.universe
```

```python
succao = ctrl.Consequent(np.arange(0,11,1),"succao")
```

```python
succao.universe
```

## Membership functions

```python
superficie.automf(number=3, names=['fácil','moderada','dificil'])
sujeira.automf(number=3, names=['leve','moderada', 'pesada'])
```

```python
superficie.view()
```

```python
superficie['fácil'].view()
```

```python
superficie['moderada'].view()
```

```python
superficie['dificil'].view()
```

```python
sujeira.view()
```

```python
sujeira['leve'].view()
```

```python
sujeira['moderada'].view()
```

```python
sujeira['pesada'].view()
```

```python
succao['baixa'] = fuzz.trimf(succao.universe,[0,0,5])
succao.view()
```

```python
succao['média'] = fuzz.trimf(succao.universe,[0,5,10])
succao.view()
```

```python
succao['alta'] = fuzz.trimf(succao.universe,[5,10,10])
succao.view()
```

```python
regra1 = ctrl.Rule(superficie['fácil'] | sujeira['leve'], succao['baixa'])
regra2 = ctrl.Rule(superficie['moderada'] | sujeira['leve'], succao['média'])
regra3 = ctrl.Rule(superficie['dificil'] | sujeira['pesada'], succao['alta'])
#regra4 = ctrl.Rule(superficie['moderada'], succao['média'])
```

## Sistema de controle

```python
#sistema_controle = ctrl.ControlSystem([regra1,regra2,regra3, regra4])
sistema_controle = ctrl.ControlSystem([regra1,regra2,regra3])
```

```python
sistema = ctrl.ControlSystemSimulation(sistema_controle)
```

```python
sistema.input['superficie'] = 4.5
sistema.input['sujeira'] = 4
sistema.compute()
```

```python
print(sistema.output['succao'])
succao.view(sim=sistema)
```

## Outras funções

```python
succao['baixa'] = fuzz.sigmf(succao.universe,3,-2)
succao['média'] = fuzz.gaussmf(succao.universe, 5, 1)
succao['alta'] = fuzz.pimf(succao.universe,5,10,10,11)
succao.view()
```

```python
regra1 = ctrl.Rule(superficie['fácil'] | sujeira['leve'], succao['baixa'])
regra2 = ctrl.Rule(superficie['moderada'] | sujeira['leve'], succao['média'])
regra3 = ctrl.Rule(superficie['dificil'] | sujeira['pesada'], succao['alta'])
#regra4 = ctrl.Rule(superficie['moderada'], succao['média'])
```

```python
#sistema_controle = ctrl.ControlSystem([regra1,regra2,regra3, regra4])
sistema_controle = ctrl.ControlSystem([regra1,regra2,regra3])
```

```python
sistema = ctrl.ControlSystemSimulation(sistema_controle)
```

```python
sistema.input['superficie'] = 4.5
sistema.input['sujeira'] = 2
sistema.compute()
```

```python
print(sistema.output['succao'])
succao.view(sim=sistema)
```
